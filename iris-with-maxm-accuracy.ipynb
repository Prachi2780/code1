**SEPARATE FEATURES AND LABELED DATA**

```python
x = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

**TRAIN - TEST SPLITTING**

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x  , y , test_size = 0.4 , random_state = 0)
print("------------FOR X_TRAIN--------------")
print(x_train)
print("------------FOR Y_TRAIN--------------")
print(y_train)
# Since all the columns have values close to each other so no feature scaling required.
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression         #-->import the model that I need to use
classifier = LogisticRegression(random_state = 0)           #-->create object of LogisticRegression class/instance of model
classifier.fit(x_train , y_train)                           #--> train on model
```

**DISPLAY THE PREDICTED VALUES**

```python
y_pred = classifier.predict(x_test)
y_pred = y_pred.reshape(len(y_pred),1)
print(y_pred)

#print(np.concatenate(y_pred.reshape(len(y_pred),1) , y_test.reshape(len(y_test),1) , 1))
```

**PREDICT A NEW VALUE**

```python
print(classifier.predict([[5.2,3.5,1.5,0.2]]))
```

**CONFUSION MATRIX AND ACCURACY**

```python
# confusion matrix is a tabular way of visualizing the performance of your prediction model
from sklearn.metrics import confusion_matrix   #-->used to evaluate the model
from sklearn.metrics import accuracy_score     #--> It gives you the overall accuracy of the model, meaning the fraction of the total samples that were correctly classified by the classifier. To calculate accuracy, use the following formula: (TP+TN)/(TP+TN+FP+FN).
cm = confusion_matrix(y_test , y_pred)         #-->there are 
print(cm)
#print(accuracy_score(np.array(x_test) , np.array(y_test)))
#print matric to get performance 
print(accuracy_score(y_test,y_pred)*100)
```

**SUPPORT VECTOR MACHINE**

```python
from sklearn.svm import SVC         #-->import the model that I need to use
classifier = SVC(kernel = 'linear' , random_state = 0)           #-->create object of LogisticRegression class/instance of model
classifier.fit(x_train , y_train)
```

**DISPLAY PREDICTED VALUES**

```python
y_pred = classifier.predict(x_test)
y_pred = y_pred.reshape(len(y_pred),1)
print(y_pred)
```

**PREDICT A NEW VALUE**

```python
print(classifier.predict([[5.2,3.5,1.5,0.2]]))
```

**CONFUSION MATRIX AND ACCURACY**

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test , y_pred)
print(cm)
#print(accuracy_score(np.array(x_test) , np.array(y_test)))
#print matric to get performance 
print(accuracy_score(y_test,y_pred)*100)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

**DISPLAY PREDICTED VALUES**

```python
y_pred = classifier.predict(x_test)
y_pred = y_pred.reshape(len(y_pred),1)
print(y_pred)
```
